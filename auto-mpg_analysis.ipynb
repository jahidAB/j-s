```python
df= df.drop(columns=['name'])
df
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
x=df.iloc[:,1:]
x
```

```python
y=df.iloc[:,:1]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,train_size=0.8)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
transformer=ColumnTransformer(transformers=[
    ('si',SimpleImputer(),['horsepower']),
    ('std',StandardScaler(),['weight']),
    ('t1',OneHotEncoder(sparse=False,drop='first'),['origin'])
],remainder='passthrough')
```

```python
a=transformer.fit_transform(x_train)
a.shape
```

```python
b=transformer.fit_transform(x_test)
b.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(a,y_train)
```

```python
pr=lr.predict(b)
pr
```

```python
lr.score(a,y_train)
```

```python
lr.score(b,y_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,pr)
mse
```

```python
import numpy as np
rmse=np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pr)
```
